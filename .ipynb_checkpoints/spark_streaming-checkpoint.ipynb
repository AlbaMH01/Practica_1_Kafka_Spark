# Spark Structured Streaming


```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType
from pyspark.sql import functions as F


spark = (
    SparkSession.builder
    .config('spark.jars.packages', 'org.apache.spark:spark-sql-kafka-0-10_2.13:3.5.0')  # Conector Kafka
    .master('local[*]')
    .appName('SportsSoccerTelemetry')
    .getOrCreate()
)
```

```python
# TODO: Define schema for Kafka messages

# Este es el "Plano" de tu mensaje
schema = StructType([
    StructField("schema_version", IntegerType(), True),
    StructField("partido_id", StringType(), True),
    StructField("jugador_id", StringType(), True),
    StructField("partido_time", StringType(), True),
    StructField("velocidad_kmh", DoubleType(), True),
    StructField("aceleracion_m_s2", DoubleType(), True),
    StructField("frecuencia_cardiaca_bpm", IntegerType(), True)
])

print("Esquema definido correctamente")
```

```python
#Kafka key juagdor_id, Timestamp 2025-12-27-T16:00:00 
{"schema_version": 1.0,
"partido_id": 1,
"jugador_id": 1,
"partido_time": "2025-12-27T16:00:00",
"velocidad_kmh": 21.94,
"aceleracion_m_s2": -0.26,
"frecuencia_cardiaca_bpm": 141
}
```

```python
# TODO: Define the topic name for Kafka producer
KAFKA_TOPIC = 'jugador.partido.metricas.v1'
```

```python
df = (
    spark.readStream
    .format('kafka')
    .option('kafka.bootstrap.servers', "localhost:9092")
    .option('subscribe', KAFKA_TOPIC)                     # Kafka topic to subscribe to
    .option('startingOffsets', 'earliest')                # Read from the beginning of the topic
    .load()
)
```
